```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file01 = './kicbase.json_df.csv'
df = pd.read_csv(file01)
```

```python
df.head(5)
```

```python
layers = ['Target', 'Severity']
df['amount'] = [1 for x in range(len(df))]
```

```python
def percentage_growth(l):
    # output:percentage of each group.
    s = 0
    res = [0]
    for i in range(len(l)-1):
        s += l[i]
        res.append(s / sum(l))
    return res
```

```python
plt.axis("equal")

cmap = plt.get_cmap("rainbow")

for i, layer in enumerate(layers):
    
    radius = i + 2 #todo
    width = 1
    frame = df.groupby(layers[:i+1])['amount'].sum()
      
    colors = cmap(percentage_growth(frame))
    labels = [x[-1] if isinstance(x, tuple) else x for x in frame.index.to_numpy()]
    plt.pie(frame, 
            labels=labels, 
            colors=colors,
            radius=radius, 
            wedgeprops=dict(width=width, edgecolor='w'), 
            labeldistance=0.8)
```
